```python
import pickle
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
from skimage.color import rgb2gray
import cv2
```

```python
# Read byte from pickle model
test_model = pickle.load(open("/content/drive/MyDrive/Classification_Model.p","rb"))
```

```python
Categories=['Apple','Jackfruit','Pomegranate','Pomelo','Tomato']
```

```python

def test(link_img):
    # Testing for a new image
  flat_data = []
  # url = input ('C:/Users/LENOVO/Downloads/062-2.jpg')
  img_array = imread(str(link_img))
  # Resize image
  img_resized = resize(img_array,(50,50,3))
  flat_data.append(img_resized.flatten())
  flat_data = np.array(flat_data)
  # print("Dimensions of original image are:",img_array.shape)

 
  y_output = test_model.predict(flat_data)
  y_output = Categories[y_output[0]]
  return 'predict '+str(y_output)
```

```python
plt.figure(figsize=(20,100))
for i in range(10):
  url='/content/sample_data/abb/'+str(i+1)+'.jpg'
  x=test(url)
  plt.subplot(20,5,i+1)
  img = plt.imread(url)
  plt.imshow(img)
  plt.title(x)
  plt.axis('off')
```
